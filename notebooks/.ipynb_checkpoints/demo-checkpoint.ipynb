# Aircraft Detection Demo

We will construct a model using a __SIFT feature extractor__ and __Gaussian process classifier with squared exponential / radial basis function (RBF) kernel__ to classify aerial optical images with 8 classes, which we narrow down to __airplane / not airplane__. Afterward, we replace the RBF kernel with a __linear kernel and a quadratic kernel__ to measure performance improvement.

Paper to improve: [Gaussian Processes for Object Detection in High Resolution Remote Sensing Images (2016)](https://ieeexplore.ieee.org/document/7838284).

Related paper: [Land Use Classification in Remote Sensing Images by Convolutional Neural Networks (2015)](https://arxiv.org/pdf/1508.00092.pdf).

## Methodology

The methods section of the paper outlines a few overall steps:

1. __Choose 8 classes__ in the land use dataset that are representative of areas around an airport. 
2. __Resize images__ to 50 x 50
3. Process the images through a __SIFT detector__ to identify keypoints.
4. Use the __Visual Bag-of-Words (VBoW)__ approach to quantize the keypoints into feature vectors.
5. Use the VBoW histograms as inputs into a __Gaussian process classifier with RBF kernel and Laplace approximation__.

## Pre-process the images for SIFT

The Scale Invariant Feature Transform (SIFT) algorithm __requires grayscale images__. In the paper, the authors also __resize the images to 50 x 50__.

```python
# import computer vision library OpenCV
import cv2
```

### Convert images to grayscale

```python
# Turn images to grayscale
r = images[:,:,:,0]
g = images[:,:,:,1]
b = images[:,:,:,2]
grayscale_images = 0.2989 * r + 0.5870 * g + 0.1140 * b
```

```python
# inspect a picture
plt.imshow(grayscale_images[0], cmap=plt.get_cmap('gray'))
plt.title(labels[0])
plt.show()
```

### Resize images to 50 x 50

```python
# create empty list to store resized images
resized_images = []

# for each image in the data set, resize to 50x50 and build new array
for img in grayscale_images:
    new_img = cv2.resize(img, (50, 50), interpolation=cv2.INTER_AREA) 
    resized_images.append(new_img)

# convert list to an array
resized_images = np.array(resized_images) 
```

```python
# inspect resized image
plt.imshow(resized_images[0], cmap=plt.get_cmap('gray'))
plt.title(labels[0])
plt.show()

# print the shape of the image
print(f"\n Shape of resized image: {resized_images[0].shape}")
```

## Extract SIFT __descriptors__

For each image, detect descriptors and store them in a list called `img_descs`.

```python
img_descs = []

for image in resized_images:
    # SIFT complains if image is not uint8
    formatted_image = (image*255).astype('uint8')
    
    # define SIFT feature extractor
    sift = cv2.xfeatures2d.SIFT_create(nOctaveLayers=60, 
                                       contrastThreshold=0.01)
    
    # extract keypoints and descriptors
    kp, desc = sift.detectAndCompute(formatted_image, None)
    
    # if no keypoints are detected, make zeros
    if len(kp) < 1:
        desc = np.zeros((1, sift.descriptorSize()), np.float32)
        
    # the descriptors should be 128 in dimensionality
    assert desc.shape[1] == 128
    
    # add descriptor to the list
    img_descs.append(desc)
    
# concatenate all of the descriptors in the dataset
image_descriptors = np.concatenate(img_descs)
```

```python
print(f"Size of the SIFT descriptors data (n_descriptors, 128): {image_descriptors.shape}")
```

## Generate SIFT __features__ with Visual Bag-of-Words (VBoW) approach

1. Using all descriptors in the dataset, __define 500 clusters__ that a descriptor in an image can fall into.
2. __Categorize each descriptor__ in an image using the cluster model
3. __Tally up the number of clusters__ in an image and store as a __histogram__.

```python
# import k-means algorithm from scikit-learn
from sklearn.cluster import MiniBatchKMeans
```

```python
# define K-means model

k_means = MiniBatchKMeans(n_clusters=500, random_state=12)

# categorize 140,082 descriptors into 500 clusters
k_means.fit(image_descriptors)

# for each descriptor, label it with the cluster it belongs to
clustered_descriptors = [k_means.predict(descriptor) for descriptor in img_descs]

# count the # of clusters in each image and create a histogram
cluster_histograms = [np.bincount(clustered_words, minlength=500) for clustered_words in clustered_descriptors]

# convert the list to an array
cluster_histograms = np.array(cluster_histograms)
```

```python
print(f"Shape of NEW and IMPROVED image dataset !! (n_images, n_clusters): {cluster_histograms.shape}")
```

## GP Classification Experiment

- __Binarize__ the target to "airplane" and "not airplane" (True or False)

- Perform __5-fold cross validation (CV)__ with the given GP + Kernel combination. 

```python
# import models from scikit-learn
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process import kernels as kern

# import helpful tools from scikit-learn
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score, roc_auc_score

# import time module and data 
import time
from pandas import DataFrame
```

```python
# BINARIZE target (airplane / not airplane)
labels = np.where(labels == 'airplane', True, False)
```

```python
# define Gaussian process with squared exponential (RBF) kernel
rbf_gp = GaussianProcessClassifier(kernel=kern.RBF(), 
                                   random_state=12)

# define Gaussian process with linear kernel
linear_gp = GaussianProcessClassifier(kernel=kern.DotProduct(sigma_0=0) + 
                                      kern.WhiteKernel(noise_level=0.001), 
                                      random_state=12)

# define Gaussian process with linear kernel
quadratic_gp = GaussianProcessClassifier(kernel=kern.DotProduct(sigma_0=0)**2 + 
                                      kern.WhiteKernel(noise_level=0.001), 
                                      random_state=12)
```

```python
def cross_validation_experiment(gp, images, labels):
    np.random.seed(12)

    # define error metrics
    training_times = []
    prediction_times = []
    accuracy_scores = []
    precision_scores = []
    recall_scores = []
    auc_scores = []
    f1_scores = []

    # define splits for 5-fold CV
    skf = StratifiedKFold(n_splits=5)
    skf.get_n_splits(images)



    # 5-fold CV loop
    for train_index, test_index in skf.split(images, labels):

        # split the data for training and evaluation
        X_train, X_test = images[train_index], images[test_index]

        y_train, y_test = labels[train_index], labels[test_index]
        
        # measure training time
        start_train_time = time.time()
        model_fit = gp.fit(X_train, y_train)
        training_time = time.time() - start_train_time

        # measure prediction time
        start_predict_time = time.time()
        yhat = gp.predict(X_test)
        prediction_time = time.time() - start_predict_time

        # calculate evaluation metrics
        accuracy = accuracy_score(y_test, yhat)
        precision = precision_score(y_test, yhat)
        recall = recall_score(y_test, yhat)
        auc = roc_auc_score(y_test, yhat)
        f1 = f1_score(y_test, yhat)

        # log results from the fold
        training_times.append(training_time)
        prediction_times.append(prediction_time)
        accuracy_scores.append(accuracy)
        precision_scores.append(precision)
        recall_scores.append(recall)
        auc_scores.append(auc)
        f1_scores.append(f1)
        
    # store the results as a dataframe
    results = DataFrame({
              'accuracy':accuracy_scores,
               'precision':precision_scores,
               'recall':recall_scores,
               #'roc auc score':auc_scores,
               #'f1 score':f1_scores,
              'training time':training_times,
              'prediction time':prediction_times
                           })
    return results
```

## Results from the paper

![](figs/paper_results.png)

```python
# run experiment for RBF GP
rbf_results = cross_validation_experiment(rbf_gp, cluster_histograms, labels)
```

```python
# run experiment for linear GP
linear_results = cross_validation_experiment(linear_gp, cluster_histograms, labels)
```

```python
# run experiment for linear GP
quadratic_results = cross_validation_experiment(quadratic_gp, cluster_histograms, labels)
```

## Results

- The __GP with RBF kernel yielded similar results__ to the __official implementation__ (lower precision and recall; it's not in the 90's)

- The GP with __quadratic kernel__ outperformed the __GP with linear kernel__ and the __GP with RBF kernel__.

- This is most likely due to the __linear decision boundary__, as seen in the Isomap projections of SIFT features.
