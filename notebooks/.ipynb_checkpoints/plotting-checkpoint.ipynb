## Import Data

Using `ImageFolder`, we can create a PyTorch `Dataset` with a simple download of the dataset from UC Merced.

The dataset has been transformed to fit the format required for ResNet-18. It is of the correct size and standardization.

However, it is not shuffled and does not load in batches.

We use `ToArray` to take a dataloader and output a NumPy array. The dataloader is helpful for batch training a neural network, but we are not training a neural network (although for subsequent experiments, we will fine-tune a neural network and then . We are only 

### PyToch DataLoader

Only for use with transfer learning on a convolutional neural network. 

```python
# keep the transforms for the dataset
transform = transforms.Compose([transforms.Resize(224),
                            transforms.CenterCrop(224),
                            transforms.ToTensor(),
                            transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                                std = [0.229, 0.224, 0.225])
                            ])

# it rubs the lotion on its skin
dataset = ImageFolder('data/', transform=transform)
dataloader = torch.utils.data.DataLoader(dataset, batch_size=32, shuffle=True, num_workers=0)


# avoid this if the dataset is large. Turns data into numpy array.
dl = ds.DataLoaderToArray()
images, labels = dl.convert(dataloader)
```
