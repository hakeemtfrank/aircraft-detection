### Visualize Image Dataset

```python
from gpcv import datasets, plotting, preprocessing
import matplotlib.pyplot as plt

## for the isomap plot
import seaborn as sns
import numpy as np
import pandas as pd


plt.style.use('seaborn')
```

```python
images, labels = datasets.load_image_data(vectorize=True, flatten=True, dim=(50,50))
```

```python
plt.imshow(images[1].reshape(50,50,3))
```

```python
from sklearn.manifold import Isomap
isomap = Isomap(n_components=2)
```

```python
image_iso = isomap.fit_transform(images)
```

```python
## plot_isomap
iso_df = pd.DataFrame({'z1':image_iso[:,0], 'z2':image_iso[:,1], 'label':labels})
iso_df['z1'] = iso_df['z1'].astype('float64')
iso_df['z2'] = iso_df['z2'].astype('float64')
```

```python
iso_df.head()
```

```python
sns.scatterplot(x=iso_df['z1'], y=iso_df['z2'], hue=iso_df['label'])
plt.show()
```

## PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
image_pca = pca.fit_transform(images)
```

```python
## plot_pca
df = pd.DataFrame({'z1':image_pca[:,0], 'z2':image_pca[:,1], 'label':labels})
df['z1'] = df['z1'].astype('float64')
df['z2'] = df['z2'].astype('float64')
```

```python
sns.scatterplot(x=df['z1'], y=df['z2'], hue=df['label'])
plt.show()
```

### TSNE

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(n_components=2, random_state=12)
image_tsne = tsne.fit_transform(images)
```

```python
df = pd.DataFrame({'z1':image_tsne[:,0], 'z2': image_tsne[:,1], 'label':labels})
df['z1'] = df['z1'].astype('float64')
df['z2'] = df['z2'].astype('float64')
```

```python
sns.scatterplot(x=df['z1'], y=df['z2'], hue=df['label'])
plt.show()
```
