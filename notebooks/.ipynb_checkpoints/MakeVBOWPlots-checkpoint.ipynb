## Gaussian process models for computer vision

- MNIST dataset
- Aerial Imagery dataset
- MRI scan dataset

My thesis investigates the effectiveness of non-stationary kernels on image tasks. My hypothesis is that:

- A special type of kernel can outperform the conventionally-used Gaussian kernel
- Feature extraction is a crucial part of image classification

Experiments are conducted to test this hypothesis. Our analysis is as follows:

1. Begin with raw images
2. Run through unsupervised feature extraction (pre-trained CNN or algorithmic feature extraction algorithm (SIFT))
3. Use 5-fold cross validation to evaluate a Gaussian process, support vector machine, logistic regression, and artificial neural network (softmax classifier). We log training and validation accuracy at each fold.

We also omit the second step to test the linear kernel on the original pixel space. For three-channel images, we convert to single-channel images using a RGB transformation (see section 2). Dimensionality reduction techniques show that there is not much information lost with this transformation. In addition, training on all three channels without a special mechanism (such as convolutions) only drastically increases processing time without any noticeable improvement with respect to training or test accuracy.


The pre-trained CNN requires the images to be at least 224x224x3 in size, with a specific mean and variance for each channel.

## SIFT Feature Extractor

```python
img_descs = []
for i in range(len(X)):
    image = (X[i,:]*255).astype('uint8')
    sift = cv2.xfeatures2d.SIFT_create(nOctaveLayers=70, contrastThreshold=0.01)
    kp, desc = sift.detectAndCompute(image, None)
    if len(kp) < 1:
        desc = np.zeros((1, sift.descriptorSize()), np.float32)
    assert desc.shape[1] == 128
    img_descs.append(desc)
X_sift = np.concatenate(img_descs)
print("{} descriptors before clustering".format(X_sift.shape[0]))
```

```python
n_clusters = 500
km_model = MiniBatchKMeans(n_clusters=n_clusters, random_state=12)
km_model.fit(X_sift)
img_clustered = [km_model.predict(raw_words) for raw_words in img_descs]
x_ = [np.bincount(clustered_words, minlength=n_clusters) for clustered_words in img_clustered]
img_bow_hist = np.array(x_)
X_hist = img_bow_hist
```

```python
def binarize(y, target):
    """
    Change multiclass to binary
    """
    y = np.where(y == target, True, False)
    return y
```

```python
y = binarize(labels, 'airplane')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_hist, y, test_size=0.2, random_state=12)
```

```python
from sklearn.svm import SVC
```

```python
model_svc = SVC()
```

```python
model_svc.fit(X_train, y_train)
```

```python
yhat = model_svc.predict(X_test)
```

```python
roc_auc_score(y_test, yhat)
```

```python
model_svc.get_params()
```

```python
model_gp = GaussianProcessClassifier(kernel=kern.RBF(length_scale_bounds=(1e-5,1)), random_state=12, n_restarts_optimizer=3)
```

```python
model_gp.fit(X_train, y_train)
```

```python
yhat_gp = model_gp.predict(X_test)
```

```python
roc_auc_score(y_test, yhat_gp)
```

```python
model_gp.kernel_
```

```python
linear_gp = GaussianProcessClassifier(kernel=kern.DotProduct(sigma_0=0.0), random_state=12, n_restarts_optimizer=1)
```

```python
linear_gp.fit(X_train, y_train)
```

```python
yhat_lingp = linear_gp.predict(X_test)
```
