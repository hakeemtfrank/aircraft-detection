- Remove the last softmax layer (the classifier layer) of the network to use a different classifier

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Pre-trained Res-net 18
resnet18 = models.resnet18(pretrained=True)

# The model for feature extraction; Remove the last layer
feature_extractor = nn.Sequential(*list(resnet18.children())[:-1])
feature_extractor.eval()
for param in feature_extractor.parameters():
    param.requires_grad = False
```

## Extract ResNet18 Feature Vectors

We also will store an appended list of our __original image batches__ so that we can compare the performance of a Gaussian process __trained on raw data__ (although converted to grayscale and resized) and a Gaussian process trained on the __embedding representations of the images__.

A method is also implemented that will transforms that vectors back to the original distribution (multiply the std and then add the mean to each channel)

```python
def pytorch_unnormalize(image_batch, mean=[0.485, 0.456, 0.406], std=[0.229,0.224,0.225]):
    """
    Takes in a PyTorch batch that has been normalized and unnormalizes it.
    https://discuss.pytorch.org/t/simple-way-to-inverse-normalize-a-batch-of-input-variable/12385
    """
    image_unnorm = image_batch.new(*image_batch.size())
    image_unnorm[:,0,:,:] = image_batch[:, 0, :, :] * std[0] + mean[0]
    image_unnorm[:,1,:,:] = image_batch[:, 1, :, :] * std[1] + mean[1]
    image_unnorm[:,2,:,:] = image_batch[:, 2, :, :] * std[2] + mean[2]
    return image_unnorm
```

```python
embedding = []
originals = []
labels_list = []

# Extract features from the images
start_time = time.perf_counter()
for images, labels in dataloader:
    image_unnorm = pytorch_unnormalize(images)
    embedding.append(feature_extractor(images))
    originals.append(image_unnorm)
    labels_list.append(labels)
extraction_duration = time.perf_counter() - start_time
print("Feature extraction time: {:.2f} minutes".format(extraction_duration / 60))
```

It takes 2 minutes to extract features with a ResNet-18 backbone.

```python
class SIFTFeatures(object):
    def __init__(self, nOctaveLayers=60, contrastThreshold=0.01, n_clusters=500):
        self.nOctaveLayers = nOctaveLayers
        self.contrastThreshold = contrastThreshold
        self.n_clusters = n_clusters
        return 
    
    def make_sift(self, X, y=None):
        """
        Generate SIFT descriptors from NumPy array. Each 

        input
        -----
        X: single-channel numpy array with pixel values from 0 to 1. 
        size: n_images, length, width (not flattened)

        parameters
        ----------
        nOctaveLayers: Number of layers per octave (default at 60)
        contrastThreshold:

        """

        img_descs = []
        for i in range(len(X)):
            image = (X[i,:]*255).astype('uint8')
            sift = cv2.xfeatures2d.SIFT_create(nOctaveLayers=self.nOctaveLayers, contrastThreshold=self.contrastThreshold)
            kp, desc = sift.detectAndCompute(image, None)
            if len(kp) < 1:
                desc = np.zeros((1, sift.descriptorSize()), np.float32)
            assert desc.shape[1] == 128
            img_descs.append(desc)
        X_descriptors = np.concatenate(img_descs)
        self.descriptors = img_descs
        return X_descriptors

    def fit(self, X, y=None):
        
        self.model_ = MiniBatchKMeans(n_clusters=self.n_clusters, random_state=12)
        X_descriptors = self.make_sift(X)
        self.model_.fit(X_descriptors)
        return self
    
    def transform(self, X, y=None):
        cluster_model = self.model_ # add self.km_model for oop. km_model should be trained with fit.

        descriptors = self.make_sift(X) # ADD self. for object implementation

        img_clustered = [cluster_model.predict(raw_words) for raw_words in self.descriptors] # add self. to img_descs for ooj self.img_descs
        x_ = [np.bincount(clustered_words, minlength=self.n_clusters) for clustered_words in img_clustered]
        X_hist = np.array(x_)
        return X_hist
```

# Visualize the data

Since we can't visualize the points near each other in d-dimensional space, we project them onto a 2 dimensional axis so that we can visualize their closeness. The quality of the visualization is dependent on the type of projection algorithm and the true "closeness" of the data in high-dimensional space. 

Each technique used below is to visualize the similarity of points. 

- Compare PCA / Isomap projections of the original data and the feature representations of the data.

## Kernel PCA

Kernel PCA is a non-linear projection technique. PCA identifies the axis that accounts for the largest amount of variance in the training set using SVD:

$$X = U\Sigma V^T$$

where $V$ contains the principal components, of which we use the first two. Denote $W_2 = [V_1 V_2]$ as the "projection matrix" that is right-multiplied to $X$ to reduce the amount of dimensions. 

Our new dataset $X \cdot W_2$ is an $n \times 2$ matrix, with each row representing the coordinates of each image on a new set of axes. The axes are selected with objective to minimize the mean squared distance between the original dataset and its projection onto the axis, or preserves the maximum amount of variance. [[1]](http://faculty.marshall.usc.edu/gareth-james/ISL/ISLR%20Seventh%20Printing.pdf)

Kernel PCA paper: [Kernel Principal Component Analysis (1999)](http://citeseerx.ist.psu.edu/viewdoc/download;jsessionid=26D9FCCBE9E9040E888E708A3BD8BAC5?doi=10.1.1.128.7613&rep=rep1&type=pdf).

Computer Vision kPCA Paper: [Kernel Principal Component Analysis and its Applications in Face Recognition and Active Shape Models](http://citeseerx.ist.psu.edu/viewdoc/download;jsessionid=26D9FCCBE9E9040E888E708A3BD8BAC5?doi=10.1.1.128.7613&rep=rep1&type=pdf).

```python
# Store the models that will be used in 5-Fold CV
# 'GPLinear', 'GP_Poly2', 'GP_Poly3',
# model_gp, model_gp2, model_gp3, 


model_names = ['GP_LinearRBF', 'GP_Linear', 'GP Poly', 'SVMLinear', 'LogisticRegression', 'GP_RBF', 'SVM_RBF']
models = [model_linear_rbf, model_linear_gp, model_poly, model_svm, model_lr, model_gprbf, model_svmrbf]
model_df = pd.DataFrame({'Model Name':model_names, 'Model':models})
```

```python
def CrossValidationTraining(model_name, X, y, k_folds=5):
    """
    Script for a single cross-validation experiment on a single model within model_df.
    Customized verison of StratifiedKFold
    """
    np.random.seed(12)
    model = model_df[model_df['Model Name'] == model_name]['Model'].to_numpy()[0]

    # Error Metrics
    training_times = []
    prediction_times = []
    accuracy_scores = []
    precision_scores = []
    recall_scores = []
    auc_scores = []
    f1_scores = []
    model_obj = model

    skf = StratifiedKFold(n_splits=k_folds, random_state=12)

    skf.get_n_splits(X)
    for train_index, test_index in skf.split(X,y):

        ## Train test split
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]

        start_train_time = time.time()
        model_fit = model.fit(X_train, y_train)
        training_time = time.time() - start_train_time

        start_predict_time = time.time()
        yhat = model_fit.predict(X_test) ## Note: may want to add predict_proba
        prediction_time = time.time() - start_predict_time

        ## Accuracy metrics
        accuracy = accuracy_score(y_test, yhat)
        precision = precision_score(y_test, yhat)
        recall = recall_score(y_test, yhat)
        auc = roc_auc_score(y_test, yhat)
        f1 = f1_score(y_test, yhat)

        ## Log fold
        training_times.append(training_time)
        prediction_times.append(prediction_time)
        accuracy_scores.append(accuracy)
        precision_scores.append(precision)
        recall_scores.append(recall)
        auc_scores.append(auc)
        f1_scores.append(f1)

    results = pd.DataFrame({'model':model_name,
              'accuracy':accuracy_scores,
               'precision':precision_scores,
               'recall':recall_scores,
               'roc auc score':auc_scores,
               'f1 score':f1_scores,
              'training time':training_times,
              'prediction time':prediction_times,
                'model object':model_obj
                           })
    return results
```

```python
def run(model_df, X, y, k_folds=5):
    """
    5-Fold Cross Validation experiment. Takes in a dataframe of models and model objects and outputs the results
    of a 5-fold CV experiment.
    
    data_name: Required; the name of the dataset (Original, SIFT, SURF)
    """
    
    np.random.seed(12)
    model_names = model_df['Model Name'].tolist()
    
    assert len(model_names) == len(set(model_names)), "Distinct models should have unique names."
    
    experiment_results = []
    for model in model_names:
        results = CrossValidationTraining(model, X, y, k_folds=k_folds)
        experiment_results.append(results)
    experiment_results = pd.concat(experiment_results)
    
    time = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    experiment_results['Experiment Run Time'] = time
    
    
    
    return experiment_results
```
