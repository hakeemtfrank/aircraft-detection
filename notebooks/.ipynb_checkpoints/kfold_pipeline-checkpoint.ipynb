## Grayscale Airplane Detection 

We will use the dataset to make an airplane detector. 

We have 2100 images when we load images using `get_landuse_data()`. 100 images per class, for 21 classes.


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import cv2 
from sklearn.model_selection import StratifiedKFold
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process import kernels as kern

from sklearn.preprocessing import LabelBinarizer, MinMaxScaler, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression

from gpcv.utils import datasets
import gpcv.preprocessing as eda
#from gpcv.models import dense_nn # keras model for testing
```

## Load Images

```python
# Load the images into memory
images, labels = datasets.get_landuse_data(classes='all', grayscale=True) # returns a list of vectors (images) and a list of strings (labels)
print("Sample size: {} images".format(len(images)))
```

```python
# Binarize the label so its just airplane. Option to balance classes
images, labels = eda.multiclass_to_binary(images, labels, target='airplane', balance=True)
```

```python
# View image chips
eda.image_matrix(image_list=images, labels=labels, dim=3)
```

```python
eda.plot_class_balance(labels)
```

## Vectorize Images

```python
# Vectorize dataset to make it ingestible 
X, y = eda.vectorize_images(images, labels)
print("Design matrix dimensions: {}".format(X.shape))
print("Label dimensions: {}".format(y.shape))
```

## Validation Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=12)
```

 ## Pipeline


```python
covariate_pipeline = Pipeline([
    ('std', StandardScaler()),
    ('mm', MinMaxScaler())
])

#label_binarizer = LabelBinarizer() # for kernas
```

```python
## Transform covariates
X_train = covariate_pipeline.fit_transform(X_train)
X_test = covariate_pipeline.transform(X_test)
```

```python
kernel = kern.RBF()
gp = GaussianProcessClassifier(kernel=kernel)
```

```python
fit = gp.fit(X_train, y_train)
yhat_gp = fit.predict(X_test)
accuracy_score(y_test, yhat_gp)
```
