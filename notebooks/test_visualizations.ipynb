This notebook is for testing the functions in `visualizations.py`. In particular, the bivariate GP sampler.

```python
import os, sys
if os.path.abspath('..') not in sys.path:
    sys.path.insert(0, os.path.abspath('..'))

from src import visualizations
from sklearn.gaussian_process import kernels as kern
import matplotlib.pyplot as plt
```

## Sample from 2D Gaussian process

```python
sine_kernel = kern.ExpSineSquared(length_scale=5, periodicity=5) + kern.DotProduct()
visualizations.plot_bivariate_gp(kernel=sine_kernel, title='Sine + Linear', 
                                 optimizer=None, granularity=60, n_samples=5,
                                random_state=12)
```

```python
rbf_fig = visualizations.plot_bivariate_gp(kernel=kern.RBF(length_scale=2), granularity=100,
                                title=None, random_state=5, n_samples=3)

cubic_fig = visualizations.plot_bivariate_gp(kernel=kern.DotProduct(sigma_0=0)**2,
                                 granularity=100, title=None, random_state=5, n_samples=3)

matern_fig = visualizations.plot_bivariate_gp(kernel=kern.Matern(), granularity=100,
                                  title=None, random_state=5, n_samples=3)
```
